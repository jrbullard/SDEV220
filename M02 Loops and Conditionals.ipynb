```python
secret = 4
guess = 6
if secret < guess:
    print("too high")
elif secret > guess:
    print("too low")
else:
    print("just right")
```

```python
guessMe = 7
number = 1
while guessMe != number :
    if number < guessMe:
        print("too low")
    if number > guessMe:
        print("oops")
        break
    else:
        print("found it!")
        break
    number += 1
```

```python
guessMe = 5
number = range(0, 10)
for num in number:
    if number < guessMe:
        print("too low")
    if number > guessMe:
        print("oops")
        break
    else:
        print("found it!")
        break
```
